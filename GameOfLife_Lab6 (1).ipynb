## The Game of Life
The Game of Life is a cellular automaton devised by the British mathematician John Horton Conway in 1970.<img align="right" src="Grid.png" HEIGHT="400" WIDTH="266"> 

The "game" is a zero-player game, meaning that its evolution is determined by its initial state, requiring no further input. One interacts with the Game of Life by creating an initial configuration and observing how it evolves.

The universe of the Game of Life is an infinite two-dimensional array of square cells, each of which is in one of two possible states:

- alive: cell has value = 1
- dead: cell has value = 0

Every cell interacts with its eight neighbours, which are the cells that are horizontally, vertically, or diagonally adjacent. See Figure 1.

## Rules of the game:

At each step in time (also called "tick"), the following transitions occur to all cells simultaneously:

1. Any live cell with fewer than two live neighbours dies, as if caused by underpopulation.

2. Any live cell with two or three live neighbours lives on to the next generation.

3. Any live cell with more than three live neighbours dies, as if by overpopulation.

4. Any dead cell with exactly three live neighbours becomes a live cell, as if by reproduction.

For the following shapes from the data folder:

Find the number of live cells and the $r_{max}$  as a function of time (number of steps).<br>

Can you determine the period of the above 3 shapes from these plots?<br>
Hint: One of the shapes does not change at all, the other 2 return back to their original configuration after certain steps.

$r_{max}$ is defined as:<br>
- Given N live cells at positions  ($i$,$j$) 
- Calculate the population "barycenter" ($center_{i}$,$center_{j}$), where $center_{i} =\frac{\sum_{i}^{N} i }{N}$ and $center_{j} =\frac{\sum_{j}^{N} j }{N}$
- Define the distance : $r_{i,j}= \sqrt{(i-center_i)^2+(j-center_j)^2}$ 
- $r_{max}$ is the maximum of all the $r_{i,j}$ i.e the distance of the furthest live cell from the "barycenter" of the population.

```python
#Import statements
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from IPython.display import HTML
import matplotlib.animation as animation
plt.rcParams["animation.html"] = "html5"

#Create function for one turn of game
#Argument is a grid of 0s and 1s
#Sums up neighbours where 1=alive and 0=dead using np.roll to go along grid
def gamestep(start):
    neighbours=np.roll(start,(1,1),axis=(0,1))+np.roll(start,(1,0),axis=(0,1))+np.roll(start,(1,-1),axis=(0,1))+np.roll(start,(-1,1),axis=(0,1))+np.roll(start,(-1,0),axis=(0,1))+np.roll(start,(-1,-1),axis=(0,1))+np.roll(start,(0,1),axis=(0,1))+np.roll(start,(0,-1),axis=(0,1))
        
    #Use bitwise or operator to set game rules
    new = ((neighbours==3)  | (start & (neighbours==2)))

    return(new)

#Create function to calculate period of game
#Runs through set number of turns and stops when original board is equal to new board
def period(start,turns):

    x=start
    turnsarray=[]
    for i in range(turns):
        y=gamestep(x)
        turnsarray.append(i)
        x=y
        
        if np.array_equal([y],[start]):
            break

    return(max(turnsarray))

#Load in data files for starting conditions
pulsar=np.loadtxt('data/Pulsar.txt',dtype=np.int8)
loaf=np.loadtxt('data/Loaf.txt',dtype=np.int8)
pentadecathlon= np.loadtxt('data/Pentadecathlon.txt',dtype=np.int8)

print ("Period of loaf is",period(loaf,20))
print ("Period of pulsar is",period(pulsar,20))
print ("Period of pentadecathlon is",period(pentadecathlon,20))


#Create array for 20 turns to be used when plotting
turns = np.arange(1,21,1)


#Create function to calculate number of alive cells for each turn of game and puts number into array
#Sums up number of cells (1=alive)
def alive(start,turns):
    livearray=[]
    for i in range(turns):
        x=np.sum(start)
        y=gamestep(start)
        livearray.append(x)
        start=y
    return (livearray)

loaflives= alive(loaf,20)
pulslives= alive(pulsar,20)
declives= alive(pentadecathlon,20)

plt.plot(turns,loaflives)
plt.title("Number of Alive Cells against Number of Steps for Loaf")
plt.xlabel("Number of Steps")
plt.ylabel("Number of Alive Cells")
plt.show()
plt.plot(turns,pulslives)
plt.title("Number of Alive Cells against Number of Steps for Pulsar")
plt.xlabel("Number of Steps")
plt.ylabel("Number of Alive Cells")
plt.show()
plt.plot(turns,declives)
plt.title("Number of Alive Cells against Number of Steps for Pentadecathlon")
plt.xlabel("Number of Steps")
plt.ylabel("Number of Alive Cells")
plt.show()


#Create function to calculate rmax
#Split up locations into rows and columns
#Calculate mean position in x and y
#Calculate max distance (r) from that mean for each cell and place into array

def rmax(start,turns):
    rmaxarray=[]
    for j in range(turns):
        board=gamestep(start)
        ri = np.where(board==1)[0]
        rj = np.where(board==1)[1]
        icentre = np.mean(ri)
        jcentre = np.mean(rj)
        r = np.sqrt((ri-icentre)**2+(rj-jcentre)**2)
        rmaxarray.append(max(r))
        start = board
    return(rmaxarray)


rloaf= rmax(loaf,20)
rpulsar= rmax(pulsar,20)
rdec = rmax(pentadecathlon,20)

plt.plot(turns,rloaf)
plt.title("rmax against Number of Steps for Loaf")
plt.xlabel("Number of Steps")
plt.ylabel("rmax")
plt.show()

plt.plot(turns,rpulsar)
plt.title("rmax against Number of Steps for Pulsar")
plt.xlabel("Number of Steps")
plt.ylabel("rmax")
plt.show()
plt.show()

plt.plot(turns,rdec)
plt.title("rmax against Number of Steps for Pentadecathlon")
plt.xlabel("Number of Steps")
plt.ylabel("rmax")
plt.show()
```

## 3: Replicator :  [3 marks]

Conway's Life is classified as following the $\textbf{B}$3/$\textbf{S}$23 rule.

Specifically a cell:
- is $\textbf{B}$orn if it has exactly 3 neighbours
- $\textbf{S}$urvives if it has 2 or 3 living neighbours 
- it dies otherwise.

Another member in the family of "Life" cellular automata is ``HighLife`` that follow the $\textbf{B}$36/$\textbf{S}$23 rule.

A cell:
- is $\textbf{B}$orn if it has 3 or 6 neighbours
- $\textbf{S}$urvives if it has 2 or 3 living neighbours 
- it dies otherwise.
